## Importing the libraries

## Task 4: ChatPromptTemplate System

```python
def get_chat_prompt(role, topic):
    system_roles = {
        "teacher": "You are a helpful teacher who explains concepts clearly.",
        "interviewer": "You are an interviewer asking insightful questions.",
        "motivator": "You are a motivational speaker who inspires learning."
    }
    chat_template = ChatPromptTemplate.from_messages([
        ("system", system_roles[role]),
        ("user", "Tell me about {topic}")
    ])
    return chat_template.format_messages(topic=topic)

messages = get_chat_prompt("teacher", "Neural Networks")
print(messages)
```

## Task 5: Input Validation Layer

```python
def validate_inputs(audience, tone):
    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]
    if audience not in valid_audience:
        audience = "beginner"
    if tone not in valid_tone:
        tone = "formal"
    return audience, tone

print(validate_inputs("kids", "fun"))
print(validate_inputs("expert", "cool"))
```

## Task 6: Prompt Generator App

```python
def generate_prompt(topic, audience, tone, style):
    audience, tone = validate_inputs(audience, tone)
    style_templates = {
        "teaching": "Explain {topic} for {audience} in a {tone} teaching style",
        "interview": "Ask interview questions about {topic} for {audience} in a {tone} tone",
        "storytelling": "Explain {topic} for {audience} in a {tone} storytelling style"
    }
    template = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=style_templates[style]
    )
    return template.format(topic=topic, audience=audience, tone=tone)

print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

## Task 7: Template Reusability Test

```python
# ONE reusable template
reusable_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

# 5 different inputs
inputs = [
    ("AI", "beginner", "fun"),
    ("Python", "intermediate", "casual"),
    ("Data Science", "expert", "formal"),
    ("ML", "beginner", "casual"),
    ("DL", "expert", "formal")
]

for topic, audience, tone in inputs:
    print(reusable_template.format(topic=topic, audience=audience, tone=tone))
```

## Pipeline Flow


```python
def run_pipeline(topic, audience, tone, style):

    print("="*60)
    print("PROMPT GENERATION PIPELINE")
    print("="*60)

    # Step 1 — User Input
    print("\nSTEP 1: User Input")
    print("Topic:", topic)
    print("Audience:", audience)
    print("Tone:", tone)
    print("Style:", style)

    # Step 2 — Input Validation
    print("\nSTEP 2: Input Validation")
    audience, tone = validate_inputs(audience, tone)
    print("Validated Audience:", audience)
    print("Validated Tone:", tone)

    # Step 3 — Prompt Template
    print("\nSTEP 3: Prompt Template Selection")

    template = PromptTemplate(
        input_variables=["topic","audience","tone","style"],
        template="Explain {topic} for {audience} in a {tone} tone using {style} style"
    )

    print("Template Structure:")
    print("Explain {topic} for {audience} in a {tone} tone using {style} style")

    # Step 4 — Prompt Generation
    print("\nSTEP 4: Dynamic Prompt Generation")

    generated_prompt = template.format(
        topic=topic,
        audience=audience,
        tone=tone,
        style=style
           )

    # Step 5 — Final Output
    print("\nSTEP 5: Final Output")
    print(generated_prompt)

    print("\nPipeline Completed Successfully")

    return generated_prompt

# Run pipeline example
run_pipeline(
    topic="Neural Networks",
    audience="beginner",
    tone="fun",
    style="storytelling"
)
```
